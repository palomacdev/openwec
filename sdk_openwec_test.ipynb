```python
import openwec

print(openwec.__version__)
```

```python
# Garante que a API está acessível
openwec.configure(base_url="http://localhost:8000/api/v1")
```

```python
# Resolve a sessão
session = openwec.Session("WEC", 2024, "Le Mans", "Race")
print(session)
```

```python
# Resultados (público, sem API key)
results = session.results()
results.head(10)
```

```python
# Voltas do carro 50 (requer API key se configurada)
laps_50 = session.laps(car="50")
laps_50.head()
```

```python
# Plot de evolução de volta
session.plot_lap_evolution(car="50")
```
